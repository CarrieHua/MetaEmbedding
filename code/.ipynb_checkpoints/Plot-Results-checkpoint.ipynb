```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

%matplotlib inline
sns.set(color_codes=True, rc={'figure.figsize':(8,6)}, 
        font_scale=1.25, style='whitegrid', font='DejaVu Serif')
```

```python
columns = [
    'logloss_base_cold', 'logloss_ME_cold', 
    'logloss_base_batcha', 'logloss_ME_batcha', 
    'logloss_base_batchb', 'logloss_ME_batchb', 
    'logloss_base_batchc', 'logloss_ME_batchc', 
    'auc_base_cold', 'auc_ME_cold', 
    'auc_base_batcha', 'auc_ME_batcha', 
    'auc_base_batchb', 'auc_ME_batchb', 
    'auc_base_batchc', 'auc_ME_batchc', 
]
methods = [
    'FM256', 'widendeep256', 'IPNN256', 'OPNN256', 'PNNstar256', 'DeepFM256'
]
res = {}
for method in methods:
    df = pd.read_csv("logs/{}.csv".format(method), 
                     names=columns, header=None)
    df["logloss_baseline"] = df["logloss_base_cold"]
    df["auc_baseline"] = df["auc_base_cold"]
    for c in columns:
        df[c+"_ratio"] = df[c]/df[c.split('_')[0]+"_baseline"] - 1
    res[method] = df
```

```python
columns_all = df.columns
get_mean_logloss_ratio = lambda method: res[method][[c for c in columns_all if 'ratio' in c and 'logloss' in c]].mean()
get_std_logloss_ratio = lambda method: res[method][[c for c in columns_all if 'ratio' in c and 'logloss' in c]].std()
get_min_logloss_ratio = lambda method: res[method][[c for c in columns_all if 'ratio' in c and 'logloss' in c]].min()
get_max_logloss_ratio = lambda method: res[method][[c for c in columns_all if 'ratio' in c and 'logloss' in c]].max()

get_mean_auc_ratio = lambda method: res[method][[c for c in columns_all if 'ratio' in c and 'auc' in c]].mean()
get_std_auc_ratio = lambda method: res[method][[c for c in columns_all if 'ratio' in c and 'auc' in c]].std()
get_min_auc_ratio = lambda method: res[method][[c for c in columns_all if 'ratio' in c and 'auc' in c]].min()
get_max_auc_ratio = lambda method: res[method][[c for c in columns_all if 'ratio' in c and 'auc' in c]].max()
```

```python
'''
Log-Loss
'''
# print("Base-model\t\t\t\t\t\t\tME")
print("\t\tcold-start", "batch-a ", "batch-b ", "batch-c ", sep="\t\t\t")
print("\t", *["Base model", "Meta-Embedding"]*4, sep=' \t')

stats = []
for method in methods:
    ans = get_mean_logloss_ratio(method)
    ans = (ans*10000).astype('int')/100.
    method = method[:-3]
    if method=='widendeep':
        method = 'Wide&Deep'
        print(method, end='\t')
    else:
        print(method, end='\t\t')
    print(*ans.tolist(), '', sep="%\t\t")
    stats.append(ans.values)
```

```python
def plot_one(ax, method, label=None, metric='auc', ylabel=True):
    label = method if label is None else label
    ans = get_mean_auc_ratio(method) if metric=='auc' else get_mean_logloss_ratio(method)
    ans_std = get_std_auc_ratio(method) if metric=='auc' else get_std_logloss_ratio(method)
    ans_min = get_min_auc_ratio(method) if metric=='auc' else get_min_logloss_ratio(method)
    ans_max = get_max_auc_ratio(method) if metric=='auc' else get_max_logloss_ratio(method)
    ans_base = ans[[i for i in ans.index if 'base' in i]].values
    ans_base_std = ans_std[[i for i in ans_std.index if 'base' in i]].values
    ans_base_min = ans_min[[i for i in ans_min.index if 'base' in i]].values
    ans_base_max = ans_max[[i for i in ans_max.index if 'base' in i]].values
    ans_me = ans[[i for i in ans.index if 'ME' in i]].values
    ans_me_std = ans_std[[i for i in ans_std.index if 'ME' in i]].values
    ans_me_min = ans_min[[i for i in ans_min.index if 'ME' in i]].values
    ans_me_max = ans_max[[i for i in ans_max.index if 'ME' in i]].values


    plt.plot(ans_me, "rD-", label="Meta-Embedding", markersize=10)
    plt.fill_between(range(4), y1=ans_me - ans_me_std, y2=ans_me + ans_me_std, alpha=0.2, color='r')
    plt.plot(ans_base, "bH-", label="Random initialized", markersize=10)
    plt.fill_between(range(4), y1=ans_base - ans_base_std, y2=ans_base + ans_base_std, alpha=0.2, color='b')

    plt.xticks(range(4),
               ["cold", "warm-a", "warm-b", "warm-c"])
    if ylabel:
        plt.ylabel("AUC improvement" if metric=='auc' else "LogLoss reduction")
    plt.xlim(0, 3.2)

    vals = ax.get_yticks()
    ax.set_yticklabels(['{:,.0%}'.format(x) for x in vals]);
    plt.title(label);
    plt.legend(loc="lower right" if metric=='auc' else "upper right", frameon=True);
```

```python
fig = plt.figure(figsize=[25, 10])
# fig = plt.figure(figsize=[10, 12/3*2])
fig.subplots_adjust(hspace=0.3, wspace=0.3)

plot_one(fig.add_subplot(2,3,1), "FM256", "FM")
plot_one(fig.add_subplot(2,3,2), "widendeep256", "Wide & Deep")
plot_one(fig.add_subplot(2,3,3), "IPNN256", "Inner PNN")
plot_one(fig.add_subplot(2,3,4), "OPNN256", "Outer PNN")
plot_one(fig.add_subplot(2,3,5), "PNNstar256", "PNN *")
plot_one(fig.add_subplot(2,3,6), "DeepFM256", "DeepFM")
```
